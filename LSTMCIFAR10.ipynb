```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.utils import to_categorical
```

```python
# Load CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Normalize pixel values to range [0, 1]
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# One-hot encode the labels (10 classes)
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Reshape images to sequences (32 time steps, 96 features)
x_train_seq = x_train.reshape(-1, 32, 96)
x_test_seq = x_test.reshape(-1, 32, 96)
```

```python
model = Sequential([
    LSTM(64, input_shape=(32, 96)),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')  # 10 output classes for CIFAR-10
])

model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    x_train_seq, y_train,
    epochs=10,
    batch_size=64,
    validation_split=0.2,
    verbose=2
)
```

```python
test_loss, test_accuracy = model.evaluate(x_test_seq, y_test, verbose=2)
print(f'Test Accuracy: {test_accuracy:.4f}')
```
